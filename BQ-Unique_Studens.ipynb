```python
#QUESTION1 
#HOW MANY UNIQUE STUDENTS ARE INCLUDED IN THE DATASET?
```

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data.xlsx")
#Select only students
Stud_data =df[df['Designation'].str.strip().str.upper().str.contains("STUDENTS")].copy()
#Email Id is the unique parameter for all students
unique_students = Stud_data['Email ID'].nunique()
print("UNIQUE STUDENTS:",unique_students)
```

```python
#Plot in pie chart
import matplotlib.pyplot as plt
import seaborn as sns
#Find non unique 
total_students = len(Stud_data)
non_unique_students = total_students -unique_students
pie_data = {
    'Category':["Unique Students","NonUnique Students"],
    'Count':[unique_students,non_unique_students]
}

plot_pie_data = pd.DataFrame(pie_data)

plt.figure(figsize=(8,8))
plt.pie(plot_pie_data['Count'],labels=plot_pie_data['Category'],autopct='%1.2f%%',startangle=100,colors=sns.color_palette('deep'))
plt.title('Pie Chart of Unique Students')
plt.show()
```
