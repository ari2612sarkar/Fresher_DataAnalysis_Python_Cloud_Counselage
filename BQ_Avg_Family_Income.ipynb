```python
#QUESTION 5
#WHAT IS THE AVERAGE FAMILY INCOME OF THE STUDENTS?
```

```python
import pandas as pd

def representative_income(income_range):
    #Remove 'lakh' from the string
    income_range = income_range.replace(' Lakh', '') 
    #Average of range of family income
    if '-' in income_range:
        lower, upper = income_range.split('-')
        return (float(lower) + float(upper)) / 2
    elif '+' in income_range:
        lower = income_range.rstrip('+')
        return float(lower) + 1
    else:
        return float(income_range)

def calculate_average_income(data):
    #Convert income ranges to representative values
    data['Income Value'] = data['Family Income'].apply(representative_income)
    
    #Calculate the average income
    avg_income = data['Income Value'].mean()
    
    return avg_income


df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data.xlsx")

#Calculate the average income
average_income = calculate_average_income(df)
print("Average Family Income:", average_income, "lakhs")
```

```python
#Histogram plot of family income
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['Family Income'], bins=10, kde=False, edgecolor='black')
plt.xlabel('Family Income')
plt.ylabel('Frequency')
plt.title('Histogram of Family Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
