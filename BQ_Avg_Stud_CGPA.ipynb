```python
#QUESTION 2
#WHAT IS THE AVERAGE GPA OF THE STUDENTS?
```

```python
import pandas as pd

def calculate_avg_cgpa(data):
    #Select Only Student Data
    stud_cgpa=data[data['Designation'].str.strip().str.upper().str.contains('STUDENTS')].copy()
    #Drop student data whose cgpa is Null
    stud_cgpa.dropna(subset=['CGPA'],inplace =True)
    stud_avg_cgpa = stud_cgpa['CGPA'].mean()
    return stud_avg_cgpa

#Extract data from excel file
df=pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data.xlsx")
avg_cgpa = calculate_avg_cgpa(df)
print(avg_cgpa)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plot CGPA of all students in a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['CGPA'].dropna(), kde=False, color='red', bins=30)

# Overlay the average CGPA as a vertical line
plt.axvline(avg_cgpa, color='black', linestyle='dashed', linewidth=2)
plt.text(avg_cgpa, plt.gca().get_ylim()[1] * 0.9, f'Average CGPA: {avg_cgpa:.2f}',  ha='center')

# Set title and labels
plt.title('Distribution of CGPA with Average CGPA Line')
plt.xlabel('CGPA')
plt.ylabel('Number of Students')

# Show plot
plt.show()
```
