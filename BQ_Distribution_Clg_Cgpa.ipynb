```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Extract data from excel file
df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data.xlsx")

#Filter data for students and drop NaN values in CGPA
stud_data = df[df['Designation'].str.strip().str.upper().str.contains("STUDENTS")].copy()
stud_data.dropna(subset=['CGPA'], inplace=True)

#Count number of unique colleges for each CGPA
unique_cgpa = stud_data.groupby('CGPA')['College Name'].nunique().reset_index()

#Bar Plot 
plt.figure(figsize=(10, 6))
sns.barplot(x='CGPA', y='College Name', data=unique_cgpa, color='red')

plt.xlabel('CGPA')
plt.ylabel('Number of Colleges')
plt.title('Number of Colleges with Unique CGPA')
plt.xticks(rotation=45)

#Adding annotations to the bars
ax = plt.gca()
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.show()
```
