```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def bar_graph_plot(data):
    #Select only student data
    stud_data = data[data['Designation'].str.upper().str.contains("STUDENTS")].copy()
    py_exp_count = stud_data['Experience with python (Months)'].value_counts().sort_index()
    
    #Plot bar graph
    plt.figure(figsize=(10,6))
    sns.barplot(x=py_exp_count.index, y=py_exp_count.values , palette='viridis')
    plt.title("Student Python Experience Distribution")
    plt.xlabel("Months of Python Experience")
    plt.ylabel("Number of Students")
    plt.show()
    
    

#Extract data from excel file
df=pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data.xlsx")
bar_graph_plot(df)
```
