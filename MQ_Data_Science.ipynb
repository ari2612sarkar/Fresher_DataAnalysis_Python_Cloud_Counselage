```python
#QUESTION 16
#FIND THE TOTAL NUMBER OF STUDENTS WHO ATTENDED THE EVENTS RELATED TO DATA SCIENCE ?
```

```python
import pandas as pd


df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data (1).xlsx")
data_event = df[df['Events'] == 'IS DATA SCIENCE FOR YOU?']

#Calculate the total number of attendees and non-attendees
total_data_attendees = len(data_event[(data_event['Designation'] == 'Students') & (data_event['Attendee Status'] == 'Attending')])
total_non_data_attendees = len(data_event[(data_event['Designation'] == 'Students') & (data_event['Attendee Status'] != 'Attending')])

print(f'Total Attendees: {total_data_attendees}')
print(f'Total Non-Attendees: {total_non_data_attendees}')

```

```python
#Data for the pie chart

import matplotlib.pyplot as plt
import seaborn as sns

labels = ['Attended', 'Did Not Attend']
sizes = [total_data_attendees, total_non_data_attendees]
colors = sns.color_palette("Set2")
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=(0.1, 0), shadow=True)
plt.title('Attendance for "IS DATA SCIENCE FOR YOU?" Event')
plt.axis('equal') 
plt.show()
```
