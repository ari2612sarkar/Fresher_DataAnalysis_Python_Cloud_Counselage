```python
import pandas as pd
df= pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data (1).xlsx")

#Group the dataframe as per events with designations
grouped_data = df.groupby(['Events', 'Designation']).size().unstack(fill_value=0)

```

```python
#Plot the KDE for 'Total Students'
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(15, 8))
sns.set_palette("pastel")
sns.kdeplot(grouped_data['Students'], shade=True, color='blue')
plt.title("KDE of Event Attendance by Event")
plt.xlabel("Number of Students")
plt.ylabel("Density")
plt.tight_layout()
```
