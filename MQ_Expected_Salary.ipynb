```python
#QUESTION10
#HOW DOES THE EXPECTED SALARY BASED ON FACTORS LIKE 'GPA','FAMILY INCOME',
#'EXPERIENCE WITH PYTHON(MONTHS)'?
```

```python
import pandas as pd

def representative_income(income_range):
    #Remove 'lakh' from the string
    income_range = income_range.replace(' Lakh', '') 
    #Average of range of family income
    if '-' in income_range:
        lower, upper = income_range.split('-')
        return (float(lower) + float(upper)) / 2
    elif '+' in income_range:
        lower = income_range.rstrip('+')
        return float(lower) + 1
    else:
        return float(income_range)

df= pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data (1).xlsx")

#Create 'Mean Income' from the range in 'Family Income' to float type
df['Mean Income'] = df['Family Income'].apply(representative_income)

#Create 'CGPA' into numeric entity
df['CGPA'] = pd.to_numeric(df['CGPA'], errors='coerce')

#Create 'Experience with python (Months)' column and convert to numeric entity
df['Experience with python (Months)'] = pd.to_numeric(df['Experience with python (Months)'], errors='coerce')

#Drop rows with null values in above columns
selected_data = df.dropna(subset=['Mean Income', 'CGPA', 'Experience with python (Months)', 'Expected salary (Lac)'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Scatterplot for Expected Salary v/s CGPA
sns.set_style("whitegrid")
plt.figure(figsize=(10, 7))
sns.scatterplot(x='CGPA', y='Expected salary (Lac)', data=selected_data, color='blue', marker='o')
plt.title('Expected Salary vs CGPA')
plt.xlabel('CGPA')
plt.ylabel('Expected salary (Lac)')
```

```python
#Sort the data by 'Family Income'
sorted_data = selected_data.sort_values(by='Mean Income')

#Scatterplot for Expected Salary v/s Family Income(Mean)
sns.set_style("whitegrid")
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Family Income', y='Expected salary (Lac)', data=sorted_data, color='green', marker='s')
plt.title('Expected Salary vs Family Income')
plt.xlabel('Family Income')
plt.ylabel('Expected salary (Lac)')
plt.show()
```

```python
#Scatterplot for Expected Salary v/s Experience with Python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Experience with python (Months)', y='Expected salary (Lac)', data=selected_data, color='red', marker='^')
plt.title('Expected Salary vs Experience with Python')
plt.xlabel('Experience with Python (Months)')
plt.ylabel('Expected salary (Lac)')
```
