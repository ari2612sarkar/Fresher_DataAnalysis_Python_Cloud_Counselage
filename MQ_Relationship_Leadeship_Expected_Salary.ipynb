```python
#QUESTION 13
#IS THERE A CORRELATION BETWEEN LEADERSHIP SKILLS AND EXPECTED SALARY OF THE STUDENTS ?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data (1).xlsx")
df['Leadership- skills'] = df['Leadership- skills'].str.strip().apply(lambda x: 1 if x.lower() == 'yes' else 0)
df['Expected salary (Lac)'] = pd.to_numeric(df['Expected salary (Lac)'], errors='coerce')  # Ensure it's numeric

#Drop rows with missing values in relevant columns
selected_data = df.dropna(subset=['Leadership- skills', 'Expected salary (Lac)'])
```

```python
#Perform linear regression using scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = linregress(selected_data['Leadership- skills'], selected_data['Expected salary (Lac)'])

#Generate points for the regression line
x_values = selected_data['Leadership- skills']
y_values = intercept + slope * x_values
```

```python
#Scatterplot and linear regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=selected_data, x='Leadership- skills', y='Expected salary (Lac)', hue='Expected salary (Lac)', palette='viridis', size='Expected salary (Lac)', sizes=(50, 200))
plt.plot(x_values, y_values, color='blue', linewidth=2, label=f'Linear Regression Line: y = {intercept:.2f} + {slope:.2f} * x')
plt.xlabel('Leadership Skills (Yes=1, No=0)')
plt.ylabel('Expected Salary (Lac)')
plt.title('Relationship between Leadership Skills and Expected Salary')
plt.legend(title='Expected Salary (Lac)', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Print the equation of the linear regression line
print("Linear Regression Equation:")
print(f"Expected Salary = {intercept:.2f} + {slope:.2f} * Leadership Skills")
```
