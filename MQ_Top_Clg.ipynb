```python
#QUESTION 18
#HOW MANY STUDENTS KNOW ABOUT THE EVENT FROM THEIR COLLEGES ? WHICH OF THESE TOP 5 COLLEGES?
```

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data (1).xlsx")

#Create Dataframe for college events 
college_events = df[df['How did you come to know about this event?'] == 'SPOC/ College Professor']

college_events = college_events['College Name'].value_counts().reset_index()
college_events.columns = ['College Name', 'Number of Students']

#Top 5 Colleges for maximun number of students
top_5_colleges = college_events.nlargest(5, 'Number of Students')
```

```python
#Bar graph plot of Top 5 Colleges
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.set_palette("pastel")
ax = sns.barplot(y='Number of Students', x='College Name', data=top_5_colleges)

plt.title("Students Who Know About the Event from Top 5 Colleges")
plt.ylabel("Number of Students")
plt.xlabel("College Name")

plt.xticks(range(len(top_5_colleges)), top_5_colleges['College Name'], rotation=45, ha='right')

plt.yticks(range(0, top_5_colleges['Number of Students'].max() + 1, 10))

for index, row in top_5_colleges.iterrows():
    plt.annotate(str(row['Number of Students']), xy=(index, row['Number of Students']),
                     ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
