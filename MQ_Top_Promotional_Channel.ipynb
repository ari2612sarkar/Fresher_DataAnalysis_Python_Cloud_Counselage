```python
#QUESTION 15
#WHICH PROMOTION CHANNEL BRINGS IN MORE STUDENT PARTICIPATIONS FOR THE EVENT?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r"C:\Users\TUF\Downloads\Data analyst Data (1).xlsx")
student_data = df[df['Designation'] == 'Students']

#Count the occurrences of each promotion channel
promotion_channel_counts = student_data['How did you come to know about this event?'].value_counts()

#Select the top 5 romotion channels
top_promotion_channels = promotion_channel_counts.head(5)
```

```python
#Bar plot and countplot
plt.figure(figsize=(12, 8))
top_promotion_channels.plot(kind='bar', color='skyblue', label='Promotion Channels')
sns.countplot(x='How did you come to know about this event?', data=student_data, order=top_promotion_channels.index, palette='viridis')
plt.xlabel('Promotion Channels')
plt.ylabel('Count')
plt.title('Top 5 Promotion Channels Among Students')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
